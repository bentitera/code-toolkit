```python
import pandas as pd
```

```python
df = pd.read_json('pitch_biggest.json')
```

```python
df.info()
```

```python
df.head()
```

```python
type(df.album[0])
```

```python
df.album = df.album.astype('str', inplace = True)
```

```python
df.album[0]
```

```python
type(df.album[0])
```

```python
df.album = df.album.str.replace('[', '').str.replace(']', '').str.replace('\'', '')
```

```python
df.head()
```

```python
df.artists = df.artists.astype('str')
```

```python
df.artists = df.artists.str.replace('[', '').str.replace(']', '').str.replace('\'', '')
```

```python
df.head()
```

```python
for col in df.columns:
    df[col] = df[col].astype('str')
    df[col] = df[col].str.replace('[', '').str.replace(']', '').str.replace('\'', '')
```

```python
df.head()
```

```python
df.score = [a[0] for a in df.score.str.split(',')]
```

```python
df.head()
```

```python
df.info()
```

```python
df.score = df.score.astype('float')
```

```python
df.score.mean()
```

```python
df.score.min()
```

```python
df.nsmallest(10, 'score')
```

```python
df.nlargest(10, 'score')
```

```python
df.time = pd.to_datetime(df.time, format='%m%d%Y', errors = 'ignore')
```

```python
df.head()
```

```python
df.info()
```

```python
def recommend():
    gen = input("What genre do you like?")
    sub = df[df.genre.str.contains(str(gen))]
    n = input("How many recommendations would you like?")
    return sub.nlargest(int(n), 'score')
```

```python
recommend()
```
